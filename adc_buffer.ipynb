## Operational amplifier (OpAmp) for instrumentation

We chose the AD8266 because of its high precision, high performance, wide voltage range, low power consumption, and overload protection, making it suitable for our application.

### Equations that govern the AD8266

These are the equations that determine the behavior of the OpAmp.

#### Output voltage
$V_{out} = G*(V_{in+}-V_{in-}) + V_{ref}$  $(1)$

#### Gain of OpAmp
$G = 1 + \frac{49.4 k\Omega}{R_G}$       $(2)$

*These equations are taken from the [datasheet.](https://www.analog.com/media/en/technical-documentation/data-sheets/AD8226.pdf)*

## Protections

Now, following the manufacturer's recommendations, power supply protections are implemented. For this purpose, two capacitors are connected in parallel to the $Vs_+$ input.

The values are $10\mu F$ and $0.1\mu F$, respectively.

Continuing with the manufacturer's recommendations, protections are implemented to protect the ADC from large voltages.

The [AD8615](https://www.analog.com/media/en/technical-documentation/data-sheets/AD8615_8616_8618.pdf) OpAmp, a $10nF$ capacitor, and two resistors, one with a value of $10k \Omega$ and the other of $10 \Omega$, were used. 

The AD8266 is an operational amplifier used as a buffer, and the AD8616 is used to scale the output between 0 and 3.3V. Then, to safeguard the integrated circuit against electrostatic discharges, TVS (Transient Voltage Suppressor) components were employed, which consist of two Zener diodes.

## Gain resistor

The following step is to calculate the value of the gain resistor among other things. For that, we will use equation 2.

```python
VADC = 3.3 #Nominal voltage of the LoRaWAN module
Vb   = 1.9 #Value above the nominal battery voltage (1.4V) to model wear and discharge
G = VADC/Vb

Rg = 49.4 /(G - 1) # Rg in kOhm

Rg = 68 #kOhms comercial value
G = 1 + 49.4/(Rg) # Gain of OpAmp 
```
